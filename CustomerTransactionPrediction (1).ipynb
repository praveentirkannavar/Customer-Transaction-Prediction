# Preprocessing

```python
df.drop('ID_code',axis=1,inplace=True)           #Unique Feature
df.head()
```

Removed unnecessary columns like ID_code is necessary beacause it has unique Values which doesn't give information to predict and ensured clean, usable data.Prepared data for modeling without missing values or noise.

# Feature & Target Split

```python
x = df.drop('target',axis=1)
y = df['target']
```

```python
x.shape
```

```python
y.shape
```

### Feature Engineering

```python
#These new features help model understand:
x["mean"] = x.mean(axis=1)                 #Overall level (mean, sum)
x["std"] = x.std(axis=1)                    #Variation (std)
x["min"] = x.min(axis=1)                  #Extremes (min, max)
x["max"] = x.max(axis=1)
x["sum"] = x.sum(axis=1)
```

```python
x["pos_count"] = (x > 0).sum(axis=1)           #POsitive Counts
x["neg_count"] = (x < 0).sum(axis=1)               #Negative Counts
x["range"] = x["max"] - x["min"]           #Range from Mx to Min
```

Created statistical features (Mean, std, Min, Max, Negative,Range etc.) to capture hidden patterns. These features improved model learning and boosted performance. Which the Input has not more information to predict.

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

Used stratified split to maintain class distribution in both sets. Ensured reliable and unbiased model evaluation. To control from imbalance we used stratify for target column.

# Preprocessing

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

To get data into one singe scale which avoids overfitting and all postive and negative values will scaled.

# Model Building

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight="balanced",C=0.01, max_iter=1000)     #Handle Imbalance
```

```python
lr = model.fit(x_train,y_train)
lr
```

# Prediction

```python
y_predict = lr.predict(x_test)
y_predict
```

```python
x_test.shape
```

```python
y_test.shape
```

# Evaluation

```python
from sklearn.metrics import roc_auc_score
lr_c = roc_auc_score(y_test, y_predict)
lr_c
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_predict))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

Used ROC-AUC as main metric due to imbalance. Also analyzed confusion matrix, precision, and recall.With this model ROC-AUC score is 78. As we see the confusion matrix that model has predicting only 0 the Customer who has transacted is not predicting .The model precision score is 30 which is very low.But FP is very low which is good.

# Problems faced 

1. Imbalanced Dataset

The dataset had a high imbalance (9:1 ratio), making the model biased towards the majority class. This affected metrics like accuracy and required special handling techniques.

2.  Computational Limitations

Large dataset size (200k rows) caused slow training and system crashes during GridSearchCV. Required reducing model complexity and using efficient methods.

3. Feature Engineering Challenges

Identifying useful features without domain knowledge was difficult. Trial-and-error was required to find features that improved performance.

4. Precision-Recall Trade-off

Improving recall led to lower precision and vice versa. Finding the right balance using threshold tuning was challenging.
